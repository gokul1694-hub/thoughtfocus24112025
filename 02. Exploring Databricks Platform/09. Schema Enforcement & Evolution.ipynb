# Schema Evolution

```python
spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_006_new_column_education.csv",
    header=True,
    inferSchema=True,
).write.option("mergeSchema","true").saveAsTable("quickstart_catalog.quickstart_schema.users_new1", mode="append")
```

```python
from pyspark.sql.functions import col
 
spark.read.table("quickstart_catalog.quickstart_schema.users_new1").filter(
    col("id") <= 4
).limit(4).display()
```

```python
spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_012_less_columns.csv",
    header=True,
    inferSchema=True,
).write.saveAsTable("quickstart_catalog.quickstart_schema.users_new", mode="append")
```

```python
from pyspark.sql.functions import col
 
spark.read.table("quickstart_catalog.quickstart_schema.users_new").filter(
    col("id") >= 5501
).limit(4).display()
```

```python
spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_013_less_columns_with_different_column_names.csv",
    header=True,
    inferSchema=True,
).select(col("user_id").alias("id"), col("full_name").alias("name")).write.saveAsTable(
    "quickstart_catalog.quickstart_schema.users_new", mode="append"
)
```
