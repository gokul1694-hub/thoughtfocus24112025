# Prequesitie
# Load users_001.csv to df

```python
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_001.csv",
    header=True,
    inferSchema=True
)
df.display()
```

# Transaction 01 - Write delta as Delta format

```python
df.write.format("delta").save(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_delta",
    mode="overwrite"
)
```

# Transaction 01 - Overwrite Delta format

```python
df.write.format("Delta").save(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_delta",
    mode="overwrite"
)
```

# Read

```python
spark.read.load(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_delta"
).display()
```

```python
from pyspark.sql.functions import col
df.filter(col("country")=='India').write.save(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_delta",
    mode='overwrite'
)
```

```python
spark.read.load(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_delta"
).display()
```

# Maintains versions

```python
spark.read.option("versionasof",0).load(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_delta"
).display()
```
