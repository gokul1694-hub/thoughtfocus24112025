# Prequisite

```python
df = spark.read.csv(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv",
                    sep="|",
                    inferSchema=True,
                    header=True,
                    quote="'")
display(df)        
```

# Dropping missing records

```python
df.na.drop(subset=["name"]).display()
```

```python
df.na.drop(subset=["name","id"]).display()
```

```python
from pyspark.sql.functions import col
df.filter(
    col("id").isNotNull() & col("name").isNotNull()
).display()
```

# Filling missing Records

```python
df.na.fill(0).display() # integer

df.na.fill(0).na.fill("NULL SOURCE").display() # string
```

```python
from pyspark.sql.functions import avg
mean = df.select(avg("exp")).collect()[0][0]
df.na.fill({
    "id":-1,
    "exp":int(mean),
    "gen":"NA"
}).display()
```
